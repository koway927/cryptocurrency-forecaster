```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Replace 'your_file_path.parquet.gz' with the path to your gzipped Parquet file
file_path1 =  './data/2024/03/13/BTC_USDT_GateIOFutures_20240313_0_orderbook_snapshot.parquet.gzip'
file_path2 =  './data/2024/03/13/BTC_USDT_GateIOFutures_20240313_0_orderbook.parquet.gzip'
file_path3 =  './data/2024/03/13/BTC_USDT_GateIOFutures_20240313_0_trades.parquet.gzip'
# Load the gzipped Parquet file
df1 = pd.read_parquet(file_path1, engine='auto')  # engine can be 'pyarrow' or 'fastparquet'
df2 = pd.read_parquet(file_path2, engine='auto')  # engine can be 'pyarrow' or 'fastparquet'
df3 = pd.read_parquet(file_path3, engine='auto')  # engine can be 'pyarrow' or 'fastparquet'
```

```python
#Data Cleaning

columns_to_convert = ['exchange_ts', 'create_ts']

dataframes = [df1, df2, df3] 
for df in dataframes: 
    df[col] = pd.to_datetime(df['capture_ts_ns'], unit='ns')# Convert 'capture_ts_ns' to datetime
    for col in columns_to_convert:
        df[col] = (df[col] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
        
```
